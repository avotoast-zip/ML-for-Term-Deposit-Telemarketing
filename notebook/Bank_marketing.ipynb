Bab ini membahas permasalahan bisnis yang dihadapi oleh sebuah bank terkait rendahnya efektivitas kampanye pemasaran deposito. Pembahasan mencakup latar belakang industri, fokus analisis, rumusan masalah, pendekatan analitik, serta metrik evaluasi yang digunakan untuk mengukur kinerja model. Tujuannya adalah menemukan solusi prediktif yang dapat meningkatkan tingkat konversi dan menekan biaya marketing.

## **1.1 Context**

Ketidaktepatan dalam menyasar target nasabah menjadi penyebab utama rendahnya efektivitas kampanye telemarketing Bank A. Banyak panggilan ditujukan kepada segmen yang tidak sesuai dengan karakteristik produk [deposito berjangka](https://www.investopedia.com/terms/t/termdeposit.asp), sehingga menghasilkan tingkat konversi yang sangat bervariasi. Pada periode tertentu, tingkat keberhasilan kampanye mencapai lebih dari 50%, namun di periode lainnya bisa turun drastis hingga di bawah 10%. Variasi ini mencerminkan bahwa strategi targeting yang digunakan belum optimal dan berdampak langsung pada tingginya biaya operasional pemasaran.

Menanggapi permasalahan tersebut, tim analisis data Bank A mengembangkan model prediktif berbasis machine learning. Model ini dibangun menggunakan data historis kampanye, mencakup atribut demografis, status keuangan, serta riwayat interaksi nasabah. Tujuannya adalah mengidentifikasi calon nasabah dengan potensi konversi lebih tinggi. Dengan pendekatan ini, tim marketing pada Bank A dapat menjalankan telemarketing secara lebih efisien, mengurangi pemborosan sumber daya, dan berpotensi meningkatkan rasio keberhasilan secara konsisten.

## **1.2 Target**

Target dari penerapan pendekatan prediktif dalam pemasaran produk perbankan, khususnya deposito berjangka, adalah untuk memperkirakan apakah seorang nasabah kemungkinan besar akan menerima atau menolak penawaran sebelum dihubungi secara langsung. Hasil prediksi ini dinyatakan dalam dua kategori, yaitu:

* "yes": nasabah bersedia dan akhirnya membuka deposito berjangka,
* "no": nasabah tidak tertarik atau menolak penawaran yang diberikan.

Dengan mengetahui kecenderungan ini sejak awal, bank dapat mengarahkan upaya pemasaran secara lebih tepat sasaran dan efisien.

## **1.3 Problem Statement**

* **Keterbatasan Anggaran Pemasaran:** Terbatasnya alokasi anggaran menyebabkan ruang gerak tim pemasaran menjadi sempit dalam menjalankan strategi yang lebih inovatif atau berskala besar. Kondisi ini menuntut setiap aktivitas promosi dijalankan secara efisien, tepat sasaran, dan berbasis data.

* **Targeting Tidak Tepat dan Minim Pemanfaatan Data**: Banyak panggilan ditujukan ke nasabah yang tidak relevan, menyebabkan pemborosan anggaran. Hal ini diperparah dengan belum optimalnya pemanfaatan data historis untuk segmentasi dan prediksi, sehingga tim marketing kesulitan menyasar nasabah dengan potensi konversi tinggi.

* **Tingkat Konversi yang Tidak Stabil:** Kampanye telemarketing menunjukkan hasil yang sangat bervariasi, mulai dari 6,43% hingga lebih dari 50% tergantung bulan. Fluktuasi ini mencerminkan kurangnya konsistensi dalam strategi pemasaran serta perlunya pendekatan yang lebih terukur dan berbasis data.

## **1.4 Goals**

Meningkatkan efektivitas kampanye telemarketing produk deposito berjangka melalui penerapan model machine learning untuk mengidentifikasi nasabah dengan potensi konversi tinggi secara lebih tepat dan efisien. Target konversi ditetapkan sebesar 20–25% per bulan, melampaui rata-rata industri perbankan sebesar 14,47% [(sumber)](https://focus-digital.co/average-sales-call-conversion-rate-by-industry/), dan mulai diukur sejak bulan pertama implementasi. Sasaran ini dinilai realistis karena konversi historis pernah mencapai hingga 50%, dan pendekatan ini juga dirancang untuk mengatasi tantangan seperti rendahnya tingkat konversi, kurangnya segmentasi nasabah, serta perlunya alokasi sumber daya yang lebih efektif dan terarah.

## **1.5 Analytic Approach**

## **1.6 Evaluation Metric**

Dalam dunia bisnis, pemahaman terhadap cost of misclassification menjadi sangat penting, yaitu biaya yang timbul akibat kesalahan dalam mengklasifikasikan target oleh model prediktif. Sebagai contoh, apabila model memprediksi bahwa seorang nasabah akan menerima tawaran produk deposito berjangka namun kenyataannya tidak, perusahaan berpotensi mengeluarkan biaya promosi yang tidak efektif. Kesalahan semacam ini dapat menimbulkan pemborosan sumber daya dan dampak negatif terhadap kinerja bisnis. Oleh karena itu, analisis dan pemahaman terhadap biaya-biaya tersebut sangat krusial untuk mengoptimalkan strategi pemasaran serta memastikan penggunaan model prediktif yang efektif dan memberikan nilai finansial yang maksimal [(advancedmartech)](https://www.advancedmartech.com/confusion-matrix/).

* **A. Oportunity Cost** <br>

    [Opportunity cost](https://www.investopedia.com/terms/o/opportunitycost.asp) adalah nilai dari peluang yang hilang ketika kita memilih satu pilihan dibandingkan pilihan lain. Dalam konteks industri perbankan, istilah ini merujuk pada potensi pendapatan yang tidak diperoleh ketika bank gagal menjangkau atau menawarkan produk deposito kepada nasabah yang sebenarnya berminat. Akibatnya, bank kehilangan kesempatan untuk mendapatkan simpanan, bunga, dan nilai bisnis dari nasabah tersebut.

    Untuk menghitung besarnya opportunity cost per nasabah deposito, berikut simulasi perhitungan berdasarkan data dan asumsi global: 

    * Total simpanan deposito di perbankan Amerika Serikat akhir Juni 2025 mencapai sekitar US $18,2 triliun [(FRED)](https://fred.stlouisfed.org/series/DPSACBW027SBOG), Jika diasumsikan 30% dari total ini berupa deposito berjangka, maka total deposito berjangka adalah: US $18,2 triliun × 30% = US $5,46 triliun
    * Asumsi Net Interest Margin (NIM) rata-rata bank global sekitar 3,2% (berdasarkan NIM industri perbankan AS Q4 2024 sebesar 3,22%) [(FDIC)](https://www.fdic.gov/news/speeches/2025/fdic-quarterly-banking-profile-fourth-quarter-2024). Maka estimasi pendapatan bank dari deposito berjangka adalah: US $5,46 triliun × 3,2% ≈ US $175 miliar
    * Jumlah rekening deposito global (estimasi kasar): Jika diasumsikan terdapat 500 juta rekening deposito aktif di seluruh dunia, maka: Pendapatan per nasabah = 

    $$
    \frac{\text{US \$175 miliar}}{500 \text{ juta nasabah}} = \boxed{\text{US \$350 per nasabah}}
    $$

    * kesimpulan : Jika satu calon nasabah deposito tidak berhasil dijangkau oleh bank, maka bank berpotensi kehilangan pendapatan bersih sekitar **US $350** per tahun dari nasabah tersebut.


* **B. Marketing Cost** <br>

    Marketing Cost adalah biaya yang dikeluarkan untuk menjangkau dan berinteraksi dengan calon nasabah deposito. Biaya ini tidak hanya mencakup tarif panggilan telepon ke berbagai operator, tetapi juga biaya infrastruktur pendukung, seperti sistem CRM, dan teknologi pendukung lainnya yang diperlukan agar tenaga telemarketing dapat bekerja dengan efisien.

    * Biaya Telepon (Outbound Telemarketing)<br>
    Menurut [Magellan Solutions](https://www.magellan-solutions.com/blog/call-center-cost-breakdown/), rata‑rata biaya layanan call center internasional (shared) adalah antara US $0,35–0,45 per menit, Dengan estimasi durasi panggilan sekitar 6 menit, dan menggunakan nilai tengah global: 6 menit × US $0,40 = US $2,40

    * Infrastruktur CRM + auto-dialer menggunakan paket SaaS seperti Salesforce. <br>
    Misalnya Salesforce Pro Suite US $100 per pengguna per bulan [(Salesforce)](https://www.salesforce.com/crm/pricing/). Diasumsikan user melakukan 500 panggilan per bulan, maka biaya per panggilan adalah: US $100 ÷ 500 = US $0,20

    * Gaji Telemarketer di Portugal <br>
    Rata-rata gaji telemarketer di Portugal adalah sekitar €1.100 per bulan [(sumber)](https://www.erieri.com/salary/job/call-center-agent/portugal), atau sekitar US $1.210 (kurs €1 = US $1,10).
    Jika seorang telemarketer melakukan 500 panggilan per bulan: US $1.210 ÷ 500 = US $2,42 per panggilan

    * Dengan demikian, estimasi Marketing Cost per panggilan/ per calon nasabah dalam skala global adalah:
    $$
    \text{US \$2,40} \text{ (Biaya Telepon)}+ \text{US \$0,2} \text{ ( Biaya Infrastruktur)} + \text{US \$2,42} \text{ ( Gaji telemarketing)} = \boxed{\text{US \$5,02 per nasabah}}
    $$


- **C. Type error**     

    Dalam memprediksi efektivitas suatu campaign marketing, memahami jenis-jenis kesalahan prediksi sangatlah penting. Kesalahan dalam memprediksi dapat merugikan pihak bank, misalnya ketika bank melakukan pendekatan kepada target atau nasabah yang tidak semestinya. Dengan memahami konsekuensi dari kesalahan ini, pihak bank dapat menyusun strategi yang lebih tepat untuk meminimalkan kerugian, meningkatkan tingkat konversi, dan mengoptimalkan keuntungan. Berikut ini adalah dua jenis kesalahan prediksi yang perlu diperhatikan :

    * False Positive (Type 1 Error)

        False Positive terjadi ketika model memprediksi bahwa seorang nasabah berpotensi tertarik pada produk deposito, tetapi dalam kenyataannya nasabah tersebut tidak tertarik atau tidak pernah melakukan pembelian. Dalam konteks ini, konsekuensi bagi bank adalah:

        * Bank harus mengeluarkan biaya marketing, termasuk biaya panggilan telepon, infrastruktur CRM, dan teknologi pendukung lainnya, untuk menjangkau nasabah yang akhirnya tidak memberikan nilai bisnis.
        * Contoh dari perhitungan di atas, bank bisa saja menghabiskan $5,2 per nasabah yang tidak pernah berkonversi. Hal ini membuat bank menanggung beban operasional yang tidak menghasilkan pendapatan, sehingga berdampak langsung pada tingkat efisiensi dan laba dari aktivitas marketing tersebut.

    * False Negative (Type 2 Error)

        False Negative terjadi ketika model memprediksi bahwa seorang nasabah tidak tertarik dengan produk deposito, padahal nasabah tersebut berpotensi atau memang siap untuk membeli produk itu. Dalam konteks ini, dampak bagi bank adalah:

        * Bank kehilangan kesempatan untuk memperoleh nilai bisnis dari nasabah tersebut, termasuk simpanan dan bunga dari deposito yang dapat digunakan untuk kebutuhan bisnis lainnya.
        * Jika estimasi nilai keuntungan dari satu nasabah deposito sebesar $350, maka setiap kesalahan ini membuat bank kehilangan potensi pendapatan sebesar nilai tersebut. Dengan kata lain, setiap kesalahan False Negative tidak hanya berarti hilangnya pendapatan, tetapi juga peluang untuk memperluas basis nasabah dan memperkuat portofolio simpanan bank.

* **D. Evaluation Metric**

    Dalam meningkatkan konversi campaign marketing, **precision merupakan metrik yang paling relevan** karena mengukur seberapa akurat prediksi prospek, yakni seberapa banyak dari nasabah yang dihubungi benar-benar tertarik. Dengan precision yang tinggi, bank dapat memfokuskan upaya dan biaya hanya pada target yang potensial, sehingga mengurangi pemborosan, meningkatkan efisiensi, dan mendorong keberhasilan campaign. Meskipun metrik seperti recall dan F1 score juga digunakan, recall cenderung menghasilkan banyak kontak tidak relevan yang memboroskan biaya, sedangkan F1 score tidak cukup spesifik untuk tujuan efisiensi konversi. Oleh karena itu, precision lebih tepat diprioritaskan agar bank dapat memaksimalkan nilai dari setiap interaksi dan meningkatkan tingkat konversi produk deposito secara optimal.


    Precision dirumuskan sebagai berikut:
    $$
    \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}
    $$
    Nilai Precision berkisar antara 0 hingga 1, di mana nilai yang lebih tinggi menunjukkan bahwa model memiliki kemampuan yang baik dalam memprediksi dengan benar nasabah yang benar-benar tertarik pada produk deposito berjangka. Dengan mengutamakan Precision, bank dapat meminimalkan biaya yang dikeluarkan untuk menargetkan nasabah yang tidak berminat, sehingga strategi pemasaran oleh tim marketing menjadi lebih efisien dan tingkat konversi meningkat. Pendekatan ini membantu mengoptimalkan sumber daya dan memastikan bahwa upaya marketing difokuskan pada calon nasabah yang paling potensial.

```python
def summary_data(df:pd.DataFrame, expand_row:bool=False) -> pd.DataFrame:
    """
    Generate summary of a DataFrame with the following details per column:
    - Data type
    - Null count and percentage
    - 'unknown' string count and percentage
    - Number of unique values
    - Sample of unique values
    """
    pd.set_option('display.max_colwidth', None if expand_row else 50)

    total_rows = len(df)

    summary = {
        "Column Name": [],
        "Type": [],
        "Null": [],
        "Null (%)": [],
        "Unknown": [],
        "Unknown (%)": [],
        "Number of Unique": [],
        "Unique Sample": []
    }

    for col in df.columns:
        col_data = df[col]
        null_count = col_data.isna().sum()
        unknown_count = (col_data == "unknown").sum()

        summary["Column Name"].append(col)
        summary["Type"].append(col_data.dtype)
        summary["Null"].append(null_count)
        summary["Null (%)"].append(round(null_count / total_rows * 100, 2))
        summary["Unknown"].append(unknown_count)
        summary["Unknown (%)"].append(round(unknown_count / total_rows * 100, 2))
        summary["Number of Unique"].append(col_data.nunique())
        summary["Unique Sample"].append(col_data.unique())

    return pd.DataFrame(summary)
```

## **3.4 Outlier Inspection**

```python
def cek_outlier_iqr(df:pd.DataFrame, num_col:list):

    result = []
    total_data = len(df)

    for col in num_col:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        L_bound = Q1 - 1.5 * IQR
        U_bound = Q3 + 1.5 * IQR

        lower_outlier = df[df[col] < L_bound]
        upper_outlier = df[df[col] > U_bound]
        total_outliers = len(lower_outlier) + len(upper_outlier)

        result.append({
                'Kolom': col,
                'Jumlah outlier lower limit': len(lower_outlier),
                'Jumlah outlier upper limit': len(upper_outlier),
                'Total outlier (%)': f'{(total_outliers / total_data) * 100:.2f}%',
                'Outlier lower limit (<=)': lower_outlier[col].max() if not lower_outlier.empty else 'None',
                'Outlier upper limit (>=)': upper_outlier[col].min() if not upper_outlier.empty else 'None'
            })
        
    return pd.DataFrame(result)
```

```python
def plot_boxplots(df, cols, ncols=2, hist_plot=False):

    nrows = (len(cols) + ncols - 1) // ncols # sama kyk roundup excel
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, nrows * 2))
    axes = axes.flatten()

    for i, col in enumerate(cols):
        if hist_plot:
            sns.histplot(df[col], ax=axes[i], kde=True, color='skyblue')
            axes[i].set_title(f'Histogram: {col}')
        else:
            sns.boxplot(x=df[col], ax=axes[i], color='skyblue')
            axes[i].set_title(f'Boxplot: {col}')
        axes[i].set_xlabel(col)

    # Matikan sumbu kosong jika ada
    for j in range(len(cols), len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def numerical_dist(df: pd.DataFrame):
    
    numerical_cols = df.select_dtypes(include=['number']).columns
    summary_dist = []

    for col in numerical_cols:

        stat, pvalue = normaltest(df[col])

        normalitas = 'Distribusi Normal' if pvalue > 0.05 else 'Distribusi Tidak Normal'

        summary_dist.append({
            'Column': col,
            "Min": df[col].min(),
            "Max": df[col].max(),
            'Mean': df[col].mean(),
            'Median': df[col].median(),
            "STD": df[col].std(),
            'p-value': f'{pvalue:.5f}',
            'Normalitas': normalitas
            })

    return pd.DataFrame(summary_dist)
```

```python
def plot_numerical_dist(df: pd.DataFrame, numerical_cols: list, size=(20, 25), bins=50):
    
    plt.figure(figsize=size)
    
    for index,col in enumerate(numerical_cols):
        # histogram
        plt.subplot(len(numerical_cols),2,index*2+1)
        sns.histplot(df_EDA[col],kde=True,color='#A5D6A7',bins=bins)
        plt.title(f'Histogram of {col}',size=12)
        plt.xlabel(col)
        plt.ylabel('Frequency')

        # boxplot
        plt.subplot(len(numerical_cols),2,index*2+2)
        sns.boxplot(x=df_EDA[col],color='skyblue')
        plt.title(f'Boxplot of {col}',size=12)
        plt.xlabel(col)

    plt.tight_layout()
    plt.show()
```

```python
def plot_numerical_EDA(df: pd.DataFrame, numerical_col: str, binary_col: str, size=(12, 2)):
    plt.figure(figsize=size)

    # Subplot 1: Boxplot
    plt.subplot(1, 2, 1)
    sns.boxplot(data=df, x=numerical_col,y=binary_col,hue=binary_col,palette='pastel',orient='h',legend=False)
    plt.title(f'Boxplot of {numerical_col} vs Target', size=12)
    plt.xlabel(numerical_col)
    plt.ylabel(binary_col)

    # Subplot 2: Barplot median
    plt.subplot(1, 2, 2)
    median = df.groupby(binary_col)[numerical_col].median().sort_index().reset_index(name='value')
    ax = sns.barplot(data=median,x='value',y=binary_col,hue=binary_col,palette='pastel',orient='h',dodge=False,legend=False)

    plt.title(f'Median {numerical_col} per Target', size=12)
    plt.xlabel(f'Median {numerical_col}')
    plt.ylabel(binary_col)

    # Tambahkan label nilai median di ujung bar
    for i, row in median.iterrows():
        ax.text(row['value'], i, f"{row['value']:.1f}", va='center', fontsize=9)

    plt.tight_layout()
    plt.show()
```

```python
def plot_categorical_pie_bar(df: pd.DataFrame, cat_col: str, size=(14, 5)):
    fig, axes = plt.subplots(1, 2, figsize=size)

    # Hitung frekuensi kategori
    count = df[cat_col].value_counts()
    labels = count.index
    sizes = count.values
    colors = sns.light_palette("#1B998B", n_colors=len(labels), reverse=True)

    # Pie chart
    axes[0].pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=140,
        colors=colors,
        textprops={'fontsize': 10, 'color': "#040404"}
    )
    axes[0].set_title(f'Proporsi {cat_col}', fontsize=12)

    # Bar chart 
    sns.barplot(
        x=sizes,
        y=labels,
        hue=labels,
        palette=colors,
        ax=axes[1],
        dodge=False,
        legend=False
    )
    axes[1].set_title(f'Jumlah per Kategori: {cat_col}', fontsize=12)
    axes[1].set_xlabel('Jumlah')
    axes[1].set_ylabel('')

    # Label jumlah di ujung bar
    for i, v in enumerate(sizes):
        axes[1].text(v + max(sizes) * 0.01, i, str(v), va='center', fontsize=12)

    plt.tight_layout()
    plt.show()
```

```python
def categorical_EDA(df: pd.DataFrame, col:str, size=(12, 5), conv_label_loc="center right", month_order=False, weekday_order=False):

    # membuat tabel data
    conv_df = df.groupby(col,observed=False).agg(
                    Yes=("y", lambda x: (x == 1).sum()),
                    No=("y", lambda x: (x == 0).sum()),
                    Total=("y", "size"),
                    Conversion_Rate=("y", lambda x: round((x == 1).sum() / len(x) * 100, 2))
                ).reset_index()
    if month_order:
        order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
            'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
        
        conv_df["month"] = pd.Categorical(conv_df["month"], categories=order, ordered=True)
        conv_df = conv_df.sort_values("month").reset_index(drop=True)
    elif weekday_order:
        order = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
        
        conv_df["day_of_week"] = pd.Categorical(conv_df["day_of_week"], categories=order, ordered=True)
        conv_df = conv_df.sort_values("day_of_week").reset_index(drop=True)

    fig, ax1 = plt.subplots(figsize=size)
    bars_yes = ax1.bar(conv_df.iloc[:,0], conv_df["Yes"], label="Yes", color="#1B998B")
    bars_no = ax1.bar(conv_df.iloc[:,0], conv_df["No"], bottom=conv_df["Yes"], label="No", color="#DAF7DC")



    # Label sumbu kiri
    ax1.set_ylabel("Telemarketing amount")
    ax1.legend(loc="upper left")

    # label total di atas bar (Yes + No)
    for i, total in enumerate(conv_df["Total"]):
        ax1.text(i, total + 100, str(total), ha='center', va='bottom', fontsize=9)

    # Line chart Conversion Rate
    ax2 = ax1.twinx()
    ax2.plot(conv_df.iloc[:,0], conv_df["Conversion_Rate"], color="#2F4858", marker="o", linewidth=2, label="Conversion Rate (%)")

    # label angka (%) line chart
    for i, (x, y) in enumerate(zip(conv_df.iloc[:,0], conv_df["Conversion_Rate"])):
        ax2.text(i+ 0.2, y + 0.1, f"{y}%", ha='center', va='bottom', fontsize=9, color="blue")
    
    # Label sumbu kanan
    ax2.set_ylabel("Conversion Rate (%)", color="blue")
    ax2.tick_params(axis='y', labelcolor="blue")
    ax2.legend(loc=conv_label_loc)


    # Tampilan akhir
    plt.title(f"Distribusi Total Telemarketing & Conversion Rate per {col}")
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()

    return conv_df
```

```python
def hipotesis_test_mannwhitneyu(df: pd.DataFrame, col: str, binary_col: str):
    
    median_no = df[df[binary_col] == 0][col]
    median_yes = df[df[binary_col] == 1][col]

    stat, p_value = mannwhitneyu(median_no, median_yes,  alternative='two-sided')

    print(f"Statistik U: {stat}, P-value: {p_value:.4f}")

    if p_value < 0.05:
        print(f"Terdapat perbedaan signifikan antara median {col} berdsarkan target.")
    else:
        print(f"Tidak terdapat perbedaan signifikan antara median {col} berdsarkan target.")
```

```python
def chi_square_test(cross_tab: pd.DataFrame, ket_col:str, alpha: float = 0.05):
    
    chi2, p, dof, expected = chi2_contingency(cross_tab)

    if p < alpha:
        print(f"P-value {p:.4f} < {alpha}, maka H₀ ditolak.")
        print(f"Terdapat hubungan yang signifikan antara {ket_col} dan keputusan terhadap penawaran deposito (y).")
    else:
        print(f"P-value {p:.4f} ≥ {alpha}, maka H₀ gagal ditolak.")
        print(f"Tidak terdapat hubungan yang signifikan antara {ket_col} dan keputusan terhadap penawaran deposito (y).")
```

## **4.5 External Factors Analysis**

Pada bab ini kita akan membahas faktor external bank seperti kondisi ekonomi, yang mungkin dapat memberikan pengaruh mengenai keputusan nasabah dalam menerima tawaran deposito.

```python
ext_df = df_EDA.groupby("month").agg(
    emp_var=('emp.var.rate', 'median'),
    num_emp=('nr.employed', 'median'),
    ihk=('cons.price.idx', 'median'),
    conf_idx=('cons.conf.idx', 'median'),
    euribor3m=('euribor3m', 'median'),
    y_yes=('y', lambda x: (x == 1).sum()),
    y_no=('y', lambda x: (x == 0).sum()),
    total=('y', 'size'),
    conv_rate=('y', lambda x: round((x == 1).sum() / len(x) * 100, 2))
).reset_index()

order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
            'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

ext_df["month"] = pd.Categorical(ext_df["month"], categories=order, ordered=True)
ext_df = ext_df.sort_values("month").reset_index(drop=True)
```

```python
def trend_ext_target(df: pd.DataFrame, col:str, size=(12, 3)):

    x = ext_df["month"]
    y1 = ext_df["conv_rate"]
    y2 = ext_df[col]

    fig, ax1 = plt.subplots(figsize=size)

    # Plot conv_rate
    ax1.plot(x, y1, color='tab:blue', marker='o', label='Conversion Rate (%)')
    # ax1.set_ylabel('Conversion Rate (%)', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    # Tambahkan label nilai di titik-titik Conversion Rate
    for i, val in enumerate(y1):
        ax1.text(i, val + 1, f"{val}%", ha='center', va='bottom', fontsize=9, color='tab:blue')
    
    # Plot emp_var di axis kanan
    ax2 = ax1.twinx()
    ax2.plot(x, y2, color='tab:red', marker='s', label=col)
    # ax2.set_ylabel(col, color='tab:red')
    ax2.tick_params(axis='y', labelcolor='tab:red')

    # Tambahkan label nilai di titik-titik Employment Variation Rate
    for i, val in enumerate(y2):
        ax2.text(i, val - 0.2, f"{val}", ha='center', va='top', fontsize=9, color='tab:red')

    # Gabungkan legend dari kedua axis
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()

    # Tempatkan legend di bawah tengah
    fig.legend(
        handles=lines1 + lines2,
        labels=labels1 + labels2,
        loc='lower center',
        bbox_to_anchor=(0.5, -0.15),
        ncol=2,
        fontsize=10
    )

    # Judul dan layout
    plt.title(f'Trend Bulanan: Conversion Rate vs {col}', fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.show()

    return ext_df[["month", col, "y_yes", "y_no", "total", "conv_rate"]]
```

### **5.1.4 Collinearity**

[VIF (Variance Inflation Factor)](https://www.investopedia.com/terms/v/variance-inflation-factor.asp) digunakan dalam analisis klasifikasi untuk mendeteksi multikolinearitas antar variabel independen. Multikolinearitas dapat menyulitkan interpretasi dan menurunkan akurasi model. Nilai VIF membantu mengidentifikasi variabel dengan korelasi tinggi agar dapat dihapus atau digabung untuk meningkatkan performa model. Interpretasinya:

* **VIF = 1**: tidak ada multikolinearitas,
* **1 < VIF ≤ 5**: korelasi moderat, masih dapat diterima,
* **VIF > 5**: indikasi multikolinearitas kuat, perlu perhatian,
* **VIF > 10**: multikolinearitas sangat kuat, perlu ditangani.

```python
# select numerical columns
num_cols_ml = df_ml.select_dtypes(include=["number"]).columns.to_list()
```

```python
def calculate_vif(df):
    
    vif_df = df.assign(constant=1)

    vif = pd.DataFrame({
        "Feature": vif_df.columns,
        "VIF": [variance_inflation_factor(vif_df.values, i) for i in range(len(vif_df.columns))]
    })
    vif = vif[vif["Feature"] != "constant"]

    return vif.sort_values("VIF", ascending=False).reset_index(drop=True)
```

```python
vif_df = calculate_vif(df_ml[num_cols_ml])
vif_df
```

```python
# menghapus dan mengulangi perhitungan VIF sampai semua nilai VIF < 10
dropped_columns = []

while True:
    vif_df = calculate_vif(df_ml[num_cols_ml].drop(columns=dropped_columns))
    
    if vif_df.iloc[0, 1] < 10:
        break
    else:
        dropped_columns.append(vif_df.iloc[0, 0])
```

### **5.1.6 Rare Values**

Rare values perlu ditangani karena dapat menyebabkan overfitting, menyulitkan interpretasi, dan menambah kompleksitas tanpa manfaat. Nilai dikatakan rare jika kemunculannya berada di bawah 5%, meskipun ambang batas ini tetap tergantung pada konteks dan domain knowledge [(sumber)](https://towardsdatascience.com/feature-engineering-examples-binning-categorical-features-9f8d582455da/).

```python
def find_rare_values(df, col, threshold=0.05):
    
    rare = df[col].value_counts(normalize=True)
    rare = rare[rare < threshold]

    result = pd.DataFrame({
        "Kolom": col,
        "Value": rare.index,
        "Frequency": rare.values
    })

    return result

def find_rare_values_in_df(df, threshold=0.05):
    
    rare_result = pd.DataFrame()

    for col in df.select_dtypes(include="object").columns:
        rare_values = find_rare_values(df, col, threshold=threshold)
        if not rare_values.empty:
            rare_result = pd.concat([rare_result, rare_values], ignore_index=True)

    return rare_result
```

#### **Experiment 1.2: Simple Imputer**

Missing value diisi menggunakan nilai median (untuk numerik) dan modus (untuk kategorikal). Pendekatan ini cocok karena mayoritas fitur tidak berkorelasi tinggi satu sama lain, dan proporsi missing cukup rendah. Namun, untuk kolom seperti `housing` dan `loan` yang memiliki pola korelasi, imputasi dilakukan berdasarkan kombinasi dominan agar tetap merepresentasikan struktur data yang ada.

```python
null_x = summary_data(x)
null_x[null_x["Null"]>0]
```

* Terlihat pada feature yang memiliki missing value semuanya adalah memiliki type object/string, oleh karena itu kita akan membuat function untuk imputasi modus
* Berdasarkan pada tahap *data cleaning* diketahui bahwa kolom `housing` dan `loan` missing valuenya saling berkorelasi sempurna (1), oleh karena itu kita akan mengisinya berdasarkan modus pada kombinasi 2 kolom tersebut.

```python
# Class modus imputer global
class ModusImputer(BaseEstimator, TransformerMixin):
    def __init__(self, fill_columns):
        self.fill_columns = fill_columns
        self.fill_values = {}

    def fit(self, X, y=None):
        # iterasi tiap kolom dan ambil modusnya
        self.fill_values = {col: X[col].mode()[0] for col in self.fill_columns}
        return self

    def transform(self, X, y=None):
        # mengisi miss val dengan dict modus
        data = X.copy()
        for col in self.fill_columns:
            data[col] = data[col].fillna(self.fill_values[col])
        return data

    def set_output(self, transform: Literal['default', 'pandas'] = 'default'):
        return super().set_output(transform=transform)
```

```python
# test code imputasi "job", "marital", "education"
ModusImputer(fill_columns=["job", "marital", "education"]).fit_transform(X_train).isna().sum()
```

```python
# Class modus imputer 2 kolom
class ModusTwoGroups(BaseEstimator, TransformerMixin):
    def __init__(self, fill_columns):
        self.fill_columns = fill_columns
        self.fill_values = {}

    def fit(self, X, y=None):
        # ambil modus kombinasi 2 kolom
        self.fill_values = ( X.groupby(self.fill_columns)
                            .size()
                            .reset_index(name='count')      # beri nama kolom frekuensi
                            .sort_values(by='count', ascending=False)  # urutkan berdasarkan frekuensi tertinggi
                            .iloc[0][self.fill_columns]   # ambil kombinasi dengan frekuensi tertinggi
                        ).to_dict()

        return self

    def transform(self, X, y=None):
        # mengisi miss val dengan dict fill_values
        data = X.copy()
        for col in self.fill_columns:
            data[col] = data[col].fillna(self.fill_values[col])
        return data

    def set_output(self, transform: Literal['default', 'pandas'] = 'default'):
        return super().set_output(transform=transform)
```

**Cross Validate**

```python
def cross_val_simple_imp(X_train, y_train, models, encoder=encoder, scaler=scaler, cv_folds=5, random_state=42):

    results = []
    cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=random_state)
    precision = make_scorer(precision_score, zero_division=0)

    for name, model in models:
        # Apply scaling only for models sensitive to feature magnitude
        if name in ['Logistic Regression', 'KNN Classifier', 'SVM Classifier']:
            pipeline = Pipeline([
                ('modus_2group', ModusTwoGroups(fill_columns=['loan', 'housing'])),
                ('modus_global', ModusImputer(fill_columns=["job", "marital", "education"])),
                ('encoder', encoder),
                ('scaler', scaler),
                ('model', model)
            ])
        else: # tree based gk pake scaler
            pipeline = Pipeline([
                ('modus_2group', ModusTwoGroups(fill_columns=['loan', 'housing'])),
                ('modus_global', ModusImputer(fill_columns=["job", "marital", "education"])),
                ('encoder', encoder),
                ('model', model)
            ])

        scores = cross_validate(
            pipeline,
            X=X_train,
            y=y_train,
            scoring=precision,
            cv=cv,
            return_train_score=False
        )

        results.append({
            'Model': name,
            'Mean Precision': np.round(scores['test_score'].mean(), 3),
            'Std Dev': np.round(scores['test_score'].std(), 3),
            'Precision Scores': np.round(scores['test_score'], 3),
            'Fit Time': np.round(scores['fit_time'].mean(), 3),
            'Score Time': np.round(scores['score_time'].mean(), 3),
        })

    return pd.DataFrame(results).sort_values(by='Mean Precision', ascending=False).reset_index(drop=True)
```

### **5.3.2 Data Imbalance Handling Experiment**

**Pada eksperimen kedua ini, fokus diarahkan pada penanganan ketidakseimbangan data (imbalance) menggunakan berbagai teknik resampling.** Beberapa metode yang akan diterapkan meliputi:

* **Over Sampling**

  * **SMOTE (Synthetic Minority Over-sampling Technique)**:  
    Membuat data sintetis dari kelas minoritas dengan menginterpolasi antar data yang ada.

  * **ADASYN (Adaptive Synthetic Sampling)**:  
    Menghasilkan data sintetis dengan fokus pada contoh yang sulit diklasifikasikan.

* **Under Sampling**

  * **Random UnderSampler**:  
    Mengurangi jumlah data dari kelas mayoritas secara acak untuk mencapai keseimbangan.

  * **NearMiss (Versi 3)**:  
    Memilih contoh dari kelas mayoritas yang paling dekat dengan data kelas minoritas.

* **Hybrid Sampling**

  * **SMOTEENN (SMOTE + Edited Nearest Neighbours)**:  
    Kombinasi antara SMOTE dan metode pembersihan data ENN, yang menghapus sampel mayoritas yang ambigu setelah oversampling.


Dengan pendekatan ini, eksperimen bertujuan untuk meningkatkan performa model pada data yang tidak seimbang.


```python
resamplers = [
    # Undersampling
    ('RandomUnderSampler', RandomUnderSampler(random_state=42)),
    ('NearMiss v3', NearMiss(version=3)),

    # Oversampling
    ('SMOTE', SMOTE(random_state=42)),
    ('ADASYN', ADASYN(random_state=42)),

    # Hybrid (SMOTE + cleaning)
    ('SMOTEENN', SMOTEENN(random_state=42))
]
```

#### **Experiment 2.1: Resample to all models**

```python
def cross_val_bal(X_train, y_train, model_name, model, resamplers=resamplers, encoder=encoder, scaler=scaler, cv_folds=5, random_state=42):

    results = []
    cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=random_state)
    scoring = {
        'precision': make_scorer(precision_score, zero_division=0),
        'recall': make_scorer(recall_score, zero_division=0),
        'f1': make_scorer(f1_score, zero_division=0)
    }

    for name, resample in resamplers:
    # Apply scaling only for models sensitive to feature magnitude
        if name in ['Logistic Regression', 'KNN Classifier', 'SVM Classifier' ]:
            pipeline = Pipeline([
                ('modus_2group', ModusTwoGroups(fill_columns=['loan', 'housing'])),
                ('modus_global', ModusImputer(fill_columns=["job", "marital", "education"])),
                ('encoder', encoder),
                ('scaler', scaler),
                ('resample', resample),
                ('model', model)
            ])
        else:  # tree-based models, no scaling
            pipeline = Pipeline([
                ('modus_2group', ModusTwoGroups(fill_columns=['loan', 'housing'])),
                ('modus_global', ModusImputer(fill_columns=["job", "marital", "education"])),
                ('encoder', encoder),
                ('resample', resample),
                ('model', model)
            ])

        scores = cross_validate(
            pipeline,
            X=X_train,
            y=y_train,
            scoring=scoring,
            cv=cv,
            return_train_score=False
        )

        results.append({
            'Model': model_name,
            'Resample': name,
            'Mean Precision': np.round(scores['test_precision'].mean(), 3),
            'Std Dev': np.round(scores['test_precision'].std(), 3),
            'Precision Scores': np.round(scores['test_precision'], 3),
            'Mean Recall': np.round(scores['test_recall'].mean(), 3),
            'Mean F1': np.round(scores['test_f1'].mean(), 3),
            'Fit Time': np.round(scores['fit_time'].mean(), 3),
            'Score Time': np.round(scores['score_time'].mean(), 3),
        })
    
    return pd.DataFrame(results).sort_values(by='Mean Precision', ascending=False).reset_index(drop=True)
```

**logistic regression** adalah metode dalam statistik dan machine learning yang digunakan untuk memprediksi kemungkinan (probabilitas) dari sebuah kejadian berdasarkan input variabel. Umumnya digunakan untuk **klasifikasi biner**, seperti memprediksi apakah pelanggan akan menerima tawaran deposito (ya/tidak), email adalah spam atau bukan, atau transaksi tergolong penipuan atau tidak.

**Cara Kerja Logistic Regression:**

Model ini menghitung **kombinasi linier** dari variabel input (fitur), kemudian menerapkan **fungsi logistik (sigmoid)** untuk mengubah hasil tersebut menjadi nilai probabilitas dalam rentang **0 hingga 1**. Probabilitas ini mencerminkan kemungkinan suatu peristiwa terjadi. Nilai probabilitas tersebut kemudian dibandingkan dengan ambang batas tertentu (umumnya 0.5) untuk menentukan kelas output:

* Jika probabilitas ≥ 0.5 (default) → diklasifikasikan sebagai **positif (1)**
* Jika probabilitas < 0.5 (default)→ diklasifikasikan sebagai **negatif (0)**

Model ini tidak hanya berguna untuk prediksi, tetapi juga untuk **mengukur pengaruh masing-masing variabel** terhadap kemungkinan terjadinya suatu hasil, karena koefisien dalam model dapat diinterpretasikan secara statistik.

### **5.4.2 Learning Curve**

[Learning curves](https://machinelearningmastery.com/learning-curves-for-diagnosing-machine-learning-model-performance/) adalah alat visual yang digunakan untuk memahami bagaimana kinerja model berubah seiring dengan jumlah data training yang digunakan. Grafik ini menggambarkan akurasi model atau error pada data pelatihan dan data validasi sebagai fungsi dari ukuran dataset. Dengan menganalisis learning curves, seseorang dapat memperoleh informasi mengenai apakah suatu model mengalami overfitting atau underfitting. Berikut merupakan kelebihan dari analisis learning curves:

- Mengidentifikasi overfitting dan underfitting:
    - Underfitting:
        - Ciri-ciri: kinerja pada data training dan validation menunjukan akurasi yang rendah dan cenderung mendatar, bahkan ketika ukuran data training meningkat. Kedua garis akan berada dekat satu sama lain yang menunjukkan model tidak mampu menangkap pola. Hal ini dapat terjadi karena model terlalu sederhana untuk memahami kompleksitas data.

    - overfitting:
        - Ciri-ciri: kinerja pada data training sangat tinggi, sedangkan pada data validation jauh lebih rendah. Garis untuk data training akan meningkat tajam, sedangkan garis pada validasi menurun setelah titik tertentu, sehingga terdapat jarak yang signifikan antara kedua garis. Hal ini dapat terjadi karena model terlalu kompleks dalam menyesuaikan diri dengan noise dalam data trainning.

**ROAS (Return on Advertising Spend)** adalah metrik yang mengukur seberapa efektif kampanye telemarketing dalam menghasilkan pendapatan. Dalam konteks bank yang menawarkan produk **deposito berjangka**, ROAS menunjukkan berapa besar pendapatan bersih yang dihasilkan dari calon nasabah yang berhasil dikonversi (True Positive), dibandingkan dengan total biaya yang dikeluarkan untuk menjangkau semua calon (baik yang berhasil maupun tidak). Nilai ROAS yang tinggi menandakan bahwa kampanye telemarketing efisien dalam menghasilkan keuntungan bagi bank. [(sumber)](https://www.advancedmartech.com/confusion-matrix/)


Berikut rumus **ROAS (Return on Ad Spend)** dalam konteks kampanye telemarketing bank untuk produk **deposito**:

**Rumus ROAS:**

$$
\text{ROAS} = \frac{(\text{TP} \times R) - (\text{FP} \times C)}{(\text{TP} + \text{FP}) \times C}
$$


**Keterangan:**

* **TP** = True Positive = jumlah nasabah yang tertarik dan benar-benar ditelepon
* **FP** = False Positive = jumlah nasabah yang tidak tertarik tetapi tetap ditelepon
* **R** = Revenue per nasabah = **US\$350**
* **C** = Biaya marketing per nasabah = **US\$5.02**



**Bussiness Simulation :**

* Biaya targeting per nasabah = **\$5.02**
* Pendapatan per nasabah berhasil (TP) = **\$350**
* Total Biaya = (TP + FP) × \$5.02 = (43+11) × \$5.02 = **\$271.08**
* Total Revenue = TP × \$350 = 43 × \$350 = **\$15,050**

**Return on Ad Spend (ROAS):**

$$
\text{ROAS} = \frac{14.994,78}{271.08} \approx \boxed{55.3}
$$


**Interpretasi:**

1. **Efisiensi Investasi Marketing Sangat Tinggi:**  
   Nilai ROAS sebesar **55.3** menunjukkan bahwa setiap **\$1** yang dikeluarkan untuk biaya telemarketing menghasilkan **\$55.3** dalam bentuk pendapatan. Ini mengindikasikan bahwa kampanye ini sangat efisien dan memberikan pengembalian yang luar biasa terhadap pengeluaran iklan atau promosi, jauh melampaui ambang wajar dalam dunia pemasaran.

**Penjelasan Koefisien dan Odds Ratio**

* **Koefisien**: Menunjukkan perubahan log-odds (logaritma dari odds) untuk setiap unit perubahan pada fitur, dengan asumsi fitur lain tetap konstan.

* **Odds Ratio (OR)**: Dihitung sebagai $e^{\text{koefisien}}$, menggambarkan perubahan **odds** (perbandingan probabilitas kejadian terhadap tidak kejadian) untuk setiap unit perubahan pada fitur.

  * **OR > 1**: Fitur meningkatkan odds dari outcome positif.
  * **OR < 1**: Fitur menurunkan odds dari outcome positif.
  * **OR = 1**: Fitur tidak mempengaruhi odds dari outcome positif. [(sumber)](https://www.geeksforgeeks.org/machine-learning/understanding-feature-importance-in-logistic-regression-models/)

Interpretasi Top 5 fitur yang paling berpengaruh:

1. **nr.employed (β = -1.3039, p < 0.001, CI \[-1.376, -1.231])**

   * Definisi: Indikator ekonomi makro – rata-rata jumlah tenaga kerja, mencerminkan kondisi pasar kerja secara umum dan bisa dikaitkan dengan daya beli masyarakat.
   * Interpretasi per 1 IQR (129 rata-rata tenaga kerja): Setiap kenaikan 129 unit rata-rata tenaga kerja, odds konversi nasabah menurun sekitar 50% (OR ≈ 0.496).
   * Interpretasi per 1 unit rata-rata tenaga kerja: Artinya setiap kenaikan 1 unit rata-rata tenaga kerja, odds konversi menurun sekitar 1% atau jika peluang konversi awalnya 70% maka akan turun menjadi 69.8%.

$$
p_0 = 0.70, \quad OR = 0.99
$$

$$
\text{odds}_0 = \frac{p_0}{1-p_0} = \frac{0.70}{0.30} = 2.33
$$

$$
\text{odds}_1 = \text{odds}_0 \times OR = 2.33 \times 0.99 = 2.31
$$

$$
p_1 = \frac{\text{odds}_1}{1+\text{odds}_1} 
= \frac{2.31}{1+2.31} \approx 0.698 \,(69.8\%)
$$

$$
\Delta p = p_1 - p_0 = 69.8\% - 70\% = -0.2\%
$$


   * Kesimpulan: Kenaikan rata-rata jumlah tenaga kerja cenderung menurunkan odds konversi nasabah, menunjukkan bahwa kondisi pasar tenaga kerja yang lebih padat mungkin membuat nasabah lebih berhati-hati dalam mengambil keputusan finansial.

## **6.2 Conclusions and Recommendations for the Model**

**Kesimpulan :**

1. **Logistic Regression dengan Simple Imputer & Threshold Default adalah Model Terbaik**  
    Logistic Regression memberikan keseimbangan optimal antara efisiensi biaya dan dampak bisnis. Model ini menghasilkan **ROAS tertinggi (55.3)** dan **conversion rate sebesar 79.6%**, menjadikannya **model paling menguntungkan secara bisnis**. Threshold default (0.5) tetap digunakan karena mampu mengonversi pelanggan lebih banyak (TP=43), meskipun threshold lebih tinggi meningkatkan efisiensi, namun mengorbankan potensi pendapatan.

2. **Penanganan Imbalance Justru Mengurangi Kinerja Model**  
    Berbagai metode resampling seperti SMOTE, ADASYN, dan class weighting **tidak berhasil meningkatkan performa model**. Precision menurun secara signifikan karena lonjakan false positive, yang berpotensi menyebabkan pemborosan biaya pemasaran. Model tanpa penanganan imbalance eksplisit justru menunjukkan performa paling stabil dan efisien.

3. **Fokus pada Precision Memastikan Efisiensi dan Minim Pemborosan**  
    Dengan precision 0.80, model sangat efektif dalam memilih nasabah yang benar-benar berminat. Ini sangat penting dalam konteks kampanye produk bernilai tinggi, di mana ketepatan targeting lebih bernilai daripada jangkauan luas.


**Rekomendasi :**

1. **Tambahkan Data yang Lebih Kaya, Berkualitas, dan Seimbang**  
   Untuk meningkatkan kapasitas prediktif model, disarankan mengintegrasikan data tambahan yang lebih informatif, seperti skor kredit, riwayat transaksi, kepemilikan produk, atau segmentasi pelanggan dari sistem CRM. Selain itu, penting untuk memastikan kualitas data yang digunakan, mulai dari kelengkapan, konsistensi, hingga pemerataan distribusi target (balance). Agar model dapat belajar dari pola yang lebih representatif dan mengurangi risiko bias prediksi.

2. **Tentukan Target yang Lebih Sesuai Kebutuhan Bisnis**  
   Saat ini model hanya memprediksi klasifikasi biner (“ya” atau “tidak”). Untuk mencerminkan tujuan bisnis secara lebih strategis, model dapat diarahkan untuk menilai tingkat minat nasabah (misalnya: rendah, sedang, tinggi). Dengan begitu, tim marketing bisa menyesuaikan strategi, apakah cukup dengan email atau perlu dihubungi langsung, sesuai potensi masing-masing nasabah.

   Perlu dibangun mekanisme pengumpulan data secara real-time dari hasil kampanye telemarketing aktual. Informasi mengenai siapa yang dihubungi, respons nasabah, dan hasil akhir kampanye dapat digunakan sebagai data baru untuk pelatihan ulang model secara berkala. Strategi ini akan meningkatkan adaptabilitas model terhadap dinamika perilaku nasabah serta menjaga relevansi model dalam jangka panjang.
